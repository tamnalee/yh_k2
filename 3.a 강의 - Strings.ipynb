### 문자열의 일부를 추출하기
- [:] 처음부터 끝까지
- [start:] start오프셋부터 끝까지
- [:end] 처음부터 end-1 오프셋까지 
- [start : end] start오프셋부터 end-1 오프셋까지
- [start : end : step] step만큼 문자를 건너뛰면서, 위와 동일하게 추출

```python
letters
```

```python
letters[0]
```

```python
# a 부터 c 까지 가져오시오.

# a ~ c

#  0 : 2  
```

```python
letters[ 0 : 2+1 ]  # 끝부분은, 내가 가져오고싶은 인덱스 숫자 + 1 
```

```python
letters[  :  6 +1 ]
```

```python
letters[ 19 : ]
```

```python
letters[ : ]
```

```python
letters[5 : 20 + 1]
```

```python
letters[ -3 :  ]
```

```python
letters
```

```python
letters[ :  : 2]  # step 자리, 몇칸 뛰어서 가져와라.
```

### 문자열 위치 찾기
- find 함수는, 찾고자 하는 문자열이 존재하는 곳의 첫번째 오프셋을 알려준다.
- rfind 함수는, 찾고자 하는 문자열이 있는 마지막 오프셋을 알려준다. 

```python
poem
```

```python
print(poem)
```

```python
# 위치 == 인덱스
```

```python
poem.find('year')
```

```python
poem.rfind('year')
```

```python
poem.find('banana')
```

```python
poem.find('So')
```
